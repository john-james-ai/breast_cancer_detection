# Metadata
CBIS-DDSM series metadata, comprising the four files below, describe the training and test sets for the calcification and mass cases.   

| # | Dataset                    | Filename                            |
|---|----------------------------|-------------------------------------|
| 1 | Mass Training Set          | mass_case_description_train_set.csv |
| 2 | Mass Test Set              | mass_case_description_test_set.csv  |
| 3 | Calcification Training Set | calc_case_description_train_set.csv |
| 4 | Calcification Test Set     | calc_case_description_test_set.csv  |

For exploration, we'll combine the training and test files.

```python
import warnings

import pandas as pd
from IPython.display import HTML, display_html

from bcd.data.prep.metadata import ColumnTransformer
from bcd.container import BCDContainer

warnings.filterwarnings("ignore")
container = BCDContainer()
container.init_resources()
```

## File Repository
Let's query the repository to get the file names

```python
repo = container.repo.meta()
repo.registry
```

Ah, there we are. 

```python
df1 = repo.get(filename='calc_case_description_test_set.csv')
df2 = repo.get(filename='mass_case_description_train_set.csv')
df3 = repo.get(filename='mass_case_description_test_set.csv')
df4 = repo.get(filename='calc_case_description_train_set.csv')
```

```python
df1.head(5)
df2.head(5)
df3.head(5)
df4.head(5)
```

Each file already has the abnormality type. We'll need to add the type of dataset. Also, there will be no need to distinguish 'BENIGN WITHOUT CALLBACK' from 'BENIGN' pathology. Let's label them all 'BENIGN'.

```python
df1["dataset"] = 'test'
df2['dataset'] = 'training'
df3['dataset'] = 'test'
df4['dataset'] = 'training'
df1.loc[df1['pathology'] == 'BENIGN_WITHOUT_CALLBACK', 'pathology'] = 'BENIGN'
df2.loc[df2['pathology'] == 'BENIGN_WITHOUT_CALLBACK', 'pathology'] = 'BENIGN'
df3.loc[df3['pathology'] == 'BENIGN_WITHOUT_CALLBACK', 'pathology'] = 'BENIGN'
df4.loc[df4['pathology'] == 'BENIGN_WITHOUT_CALLBACK', 'pathology'] = 'BENIGN'
```

Ok, let's see what we have.

```python
df1.head(5)
df2.head(5)
df3.head(5)
df4.head(5)
```

Viola, let's combine the training and test files, and commit them to the repository.

```python
calc = pd.concat([df1,df4], axis=0)
mass = pd.concat([df2,df3], axis=0)
```

Almost done. Some of the column names have spaces in them. Let's replace them with underscores to avoid downstream vaxations.

```python
ct = ColumnTransformer()
calc = ct.fit_transform(calc)
mass = ct.fit_transform(mass)
```

ok, let's check the column names

```python
calc.columns
mass.columns
```

Viola! Save!

```python
repo.add(filename="calc_cases.csv", data=calc, force=True)
repo.add(filename='mass_cases.csv', data=mass, force=True)
```

Done.
